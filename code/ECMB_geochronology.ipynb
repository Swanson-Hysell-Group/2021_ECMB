## U-Pb zircon geochronology

### ECMB6 (St. Cloud Granite)
**Sample**: ECMB-6  
**Unit**: St. Cloud Granite  
**Lat**: 45.53396  
**Lon**: -94.23187  
**Elevation**: 345 m  
**Description**: The St. Cloud Granite was emplaced into the Reformatory Granodiorite and is well-exposed throughout Stearns County Quarry Park. This sample St. Cloud Red Granite was collected from a fractured pavement outcrop. This pink hornblende-bearing granite has microcline that is typically 4-8 mm and plag that is 4-6 mm.

```python
columns = np.arange(0,23)

ECMB6 = pd.read_excel('../data/geochron/ECMB_FDT_19Nov20.xlsx',
                      header=26,nrows=8,usecols=columns,
                      names=['zircon','Th_U','206Pb_mol','206Pb_mol%','Pb*/Pbc',
                                              'Pbc','206Pb/204Pb','','','','','','','','','','',
                                              '207Pb/206Pb_date','207Pb/206Pb_date_2sigma',
                                              '207Pb/235U_date','207Pb/235U_date_2sigma',
                                              '206Pb/238U_date','206Pb/238U_date_2sigma'])
ECMB6['include'] = pd.Series(data=['yes','yes','yes','yes','yes','yes','yes','yes'])

ECMB6.sort_values('207Pb/206Pb_date', inplace=True, ascending=False)
ECMB6.reset_index(inplace=True)
```

```python
ECMB6
```

```python
ECMB6_mean, ECMB6_2sigma = weighted_mean(ECMB6['207Pb/206Pb_date'].tolist(),ECMB6['207Pb/206Pb_date_2sigma'].tolist())
```

### ECMB4 (Richmond Granite)

**Sample**: ECMB-4  
**Unit**: Richmond Granite (1772 Â± 3; Holm et al. 2005)  
**Lat**: 45.443429  
**Lon**: -94.483603  
**Elevation**: 333 m  
**Description**: Abandoned quarry outcrop on the side of extensive hilltop exposure of the Richmond Granite. The granite has large k-spar phenocrysts (typically 2 cm in length) with smaller orthopyroxene and quartz. The granite has charnockitic affinity given the presence of orthopyroxene.

```python
columns = np.arange(0,23)

ECMB4 = pd.read_excel('../data/geochron/ECMB_FDT_19Nov20.xlsx',
                      header=6,nrows=8,usecols=columns,
                      names=['zircon','Th_U','206Pb_mol','206Pb_mol%','Pb*/Pbc',
                                              'Pbc','206Pb/204Pb','','','','','','','','','','',
                                              '207Pb/206Pb_date','207Pb/206Pb_date_2sigma',
                                              '207Pb/235U_date','207Pb/235U_date_2sigma',
                                              '206Pb/238U_date','206Pb/238U_date_2sigma'])
ECMB4['include'] = pd.Series(data=['yes','yes','yes','yes','yes','yes','no','yes'])

ECMB4.sort_values('207Pb/206Pb_date', inplace=True, ascending=False)
ECMB4.reset_index(inplace=True)
```

```python
ECMB4
```

```python
ECMB4_filtered = ECMB4[ECMB4['include']=='yes']
ECMB4_mean, ECMB4_2sigma = weighted_mean(ECMB4_filtered['207Pb/206Pb_date'].tolist(),ECMB4_filtered['207Pb/206Pb_date_2sigma'].tolist())
```

```python
U_Pb_zircon_dates = [ECMB6['207Pb/206Pb_date'].tolist(),
                    QP1['207Pb/206Pb_date'].tolist(),
                    ECMB4['207Pb/206Pb_date'].tolist()]

U_Pb_zircon_2sigmas = [ECMB6['207Pb/206Pb_date_2sigma'].tolist(),
                      QP1['207Pb/206Pb_date_2sigma'].tolist(),
                      ECMB4['207Pb/206Pb_date_2sigma'].tolist()]

U_Pb_zircon_include = [ECMB6['include'].tolist(),
                       QP1['include'].tolist(),
                       ECMB4['include'].tolist()]

U_Pb_sample_means = [ECMB6_mean,
                     QP1_mean,
                     ECMB4_mean]

U_Pb_sample_2sigmas = [ECMB6_2sigma, 
                       QP1_2sigma, 
                       ECMB4_2sigma]

colors = ['tomato',
          '#D98880',
          'orange']

labels = ['St. Cloud Granite',
          'Quartz-Porphyry Dike',
          'Richmond Granite']

sample_ID = ['ECMB6',
             'QP1',
             'ECMB4']

latitude = [45.53396,
            45.53481,
            45.443429]

longitude = [-94.23187,
             -94.25811,
             -94.483603]
```

```python
ECMB_geochron = pd.DataFrame({'sample_ID':sample_ID,
                              'labels': labels, 
                              'polarity_colors' : colors, 
                              'U_Pb_zircon_dates': U_Pb_zircon_dates, 
                              'U_Pb_zircon_2sigmas': U_Pb_zircon_2sigmas,
                              'U_Pb_zircon_include': U_Pb_zircon_include,
                              'U_Pb_sample_mean': U_Pb_sample_means,
                              'U_Pb_sample_2sigma': U_Pb_sample_2sigmas,
                              'latitude':latitude,
                              'longitude':longitude})

ECMB_geochron.to_csv('../data/geochron/compiled_geochron.csv')
```

```python
ECMB6_age_str = str(np.round(ECMB6_mean,2)) + 'Â±' + str(np.round(ECMB6_2sigma,2)) + ' Ma'
QP1_age_str = str(np.round(QP1_mean,2)) + 'Â±' + str(np.round(QP1_2sigma,2)) + ' Ma'
ECMB4_age_str = str(np.round(ECMB4_mean,2)) + 'Â±' + str(np.round(ECMB4_2sigma,2)) + ' Ma'
```

```python
dike_max_age = ECMB6_mean+ECMB6_2sigma
dike_min_age = ECMB4_mean-ECMB4_2sigma
dike_mid_point_age = (dike_max_age+dike_min_age)/2
```

```python
fig, ax, x_positions = plot_dates(ECMB_geochron.U_Pb_zircon_dates, ECMB_geochron.U_Pb_zircon_2sigmas, 
                                  ECMB_geochron.U_Pb_zircon_include, 
                                  ECMB_geochron.U_Pb_sample_mean, ECMB_geochron.U_Pb_sample_2sigma, 
                                  colors, labels, legend=False, figsize=(12,6), lineweight=6,
                                  ylabel = '$^{207}$Pb/$^{206}$Pb date (Ma)')

max_x_value = 40
ax.set_xlim(0,max_x_value)
ax.set_xticks([])

ax.set_ylim(1786,1771)
ax.tick_params(axis='both', which='major', labelsize=12)

ax.text(x_positions[0]-4,ECMB6_mean-3,'ECMB6 \n$\mathit{St.\;Cloud\;Granite}$\n'+ECMB6_age_str,size='14',horizontalalignment='center',
         verticalalignment='bottom',multialignment='center')
ax.text(x_positions[1]-4,QP1_mean-3,'QP1 \n$\mathit{qtz-porphyry\;dike}$\n'+QP1_age_str,size='14',horizontalalignment='center',
         verticalalignment='bottom',multialignment='center')
ax.text(x_positions[2]-5,ECMB4_mean-3,'ECMB4 \n$\mathit{Richmond\;Granite}$\n'+ECMB4_age_str,size='14',horizontalalignment='center',
         verticalalignment='bottom',multialignment='center')

dike_bar_x = 33
ax.vlines(dike_bar_x,ECMB6_mean+ECMB6_2sigma,ECMB4_mean-ECMB4_2sigma,color='grey',linewidth=5)
ax.hlines(ECMB6_mean+ECMB6_2sigma,dike_bar_x-0.5,dike_bar_x+0.5,color='grey',linewidth=5)
ax.hlines(ECMB4_mean-ECMB4_2sigma,dike_bar_x-0.5,dike_bar_x+0.5,color='grey',linewidth=5)
ax.text(dike_bar_x+3.5,dike_mid_point_age,'NED diabase\ndike age range\n \n1776.3 Ma\nto\n1781.9 Ma',size='14',horizontalalignment='center',
         verticalalignment='center',multialignment='center')

legend_elements = [Patch(facecolor='lightgrey', edgecolor=None,
                         label='weighted mean dates'),
                   Line2D([], [],  marker='|', linestyle='None', color='tomato', label='individual zircon dates',
                          markerfacecolor='k', markersize=10, markeredgewidth=3)]

plt.legend(handles=legend_elements, loc='upper right',
           bbox_to_anchor=(.31,1), fancybox=True, fontsize = 13)

plt.savefig('../Manuscript/Figures/ECMB_new_U_Pb_dates.pdf',bbox_inches='tight')
```

## Thermochronology

Constraints on the thermochronology of the ECMB come from Ar-Ar on hornblende and micas developed by Holm et al. (2005) as well as new U-Pb apatite data from this study.

```python
Holm_U_Pb = pd.read_csv('../data/geochron/Holm2005_U_Pb_dates.csv')
Holm_U_Pb.sort_values('U-Pb age', inplace=True, ascending=True)
```

```python
Holm_U_Pb_filtered = Holm_U_Pb[Holm_U_Pb['2sigma']<15]
```

```python
Holm_Ar_Ar = pd.read_csv('../data/geochron/Holm_Ar_Ar_dates.csv')
Holm_Ar_Ar.sort_values('Ar-Ar age', inplace=True, ascending=True)
Holm_Ar_Ar_ECMB = Holm_Ar_Ar[Holm_Ar_Ar.Unit == 'ECMB']
Holm_Ar_Ar_preECMB = Holm_Ar_Ar[Holm_Ar_Ar.Unit == 'pre-ECMB']
```

```python
Holm_Ar_Ar_ECMB_biotite = Holm_Ar_Ar_ECMB[Holm_Ar_Ar_ECMB['mineral dated'] == 'biotite']
Holm_Ar_Ar_ECMB_hornblende = Holm_Ar_Ar_ECMB[Holm_Ar_Ar_ECMB['mineral dated'] == 'hornblende']
Holm_Ar_Ar_preECMB_biotite = Holm_Ar_Ar_preECMB[Holm_Ar_Ar_preECMB['mineral dated'] == 'biotite']
```

```python
Holm_Ar_Ar_ECMB_hornblende
```

```python
#new ECMB apatite dates
U_Pb_apatite_dates = [1751.67,1800.3,1810.54]
U_Pb_apatite_2sigma = [36.55,65.2,22.95]
```

```python
n_biotite = len(Holm_Ar_Ar_ECMB_biotite)
n_hornblende = len(Holm_Ar_Ar_ECMB_hornblende)
n_apatite = len(U_Pb_apatite_dates)
n_zircon = len(Holm_U_Pb_filtered)
n_zircon_new = len(ECMB_geochron)

fig = plt.figure(figsize=(9.3,6))
ax = plt.subplot()

plt.errorbar(Holm_Ar_Ar_preECMB_biotite['Ar-Ar age'],np.arange(-len(Holm_Ar_Ar_preECMB_biotite)-2,-2),xerr=Holm_Ar_Ar_preECMB_biotite['2sigma'],fmt='>',
            label='pre-ECMB Ar-Ar biotite dates\n(Holm and Lux, 1996; Holm et al., 2005)',color='black')

plt.errorbar(Holm_Ar_Ar_ECMB_biotite['Ar-Ar age'],np.arange(0,n_biotite),xerr=Holm_Ar_Ar_ECMB_biotite['2sigma'],fmt='s',
            label='ECMB Ar-Ar biotite dates\n(Holm and Lux, 1996; Holm et al., 2005)')

plt.errorbar(Holm_Ar_Ar_ECMB_hornblende['Ar-Ar age'],np.arange(n_biotite,n_biotite+n_hornblende),
             xerr=Holm_Ar_Ar_ECMB_hornblende['2sigma'],fmt='h',
             label='ECMB Ar-Ar hornblende dates\n(Holm et al., 2005)')

plt.errorbar(U_Pb_apatite_dates,np.arange(n_biotite+n_hornblende,n_biotite+n_hornblende+n_apatite),
             xerr=U_Pb_apatite_2sigma,fmt='o',
            label='ECMB U-Pb apatite dates\n(this study)')

plt.errorbar(Holm_U_Pb_filtered['U-Pb age'],np.arange(n_biotite+n_hornblende+n_apatite,n_biotite+n_hornblende+n_apatite+n_zircon),
             xerr=Holm_U_Pb_filtered['2sigma'],fmt='D',
            label='ECMB U-Pb zircon dates\n(Holm et al., 2005)')

plt.errorbar(ECMB_geochron.U_Pb_sample_mean, np.arange(n_biotite+n_hornblende+n_apatite+n_zircon,n_biotite+n_hornblende+n_apatite+n_zircon+n_zircon_new),
             xerr=[2.4,2.4,2.4],fmt='d',
            label='ECMB U-Pb zircon dates\n(this study)',c='tomato')

ax.text(1710,(n_biotite+n_hornblende+n_apatite+n_biotite+n_hornblende+n_apatite+n_zircon+n_zircon_new)/2,'ECMB: U-Pb zircon (crystallization)',size='11',horizontalalignment='left',
         verticalalignment='center',multialignment='center',color='C3')

ax.text(1710,((n_biotite+n_hornblende)+(n_biotite+n_hornblende+n_apatite))/2,'ECMB: U-Pb apatite (~550-450Â°C)',size='11',horizontalalignment='left',
         verticalalignment='center',multialignment='center',color='C2')

ax.text(1710,(n_biotite+(n_biotite+n_hornblende))/2,'ECMB: Ar-Ar hornblende (~500Â°C)',size='11',horizontalalignment='left',
         verticalalignment='center',multialignment='center',color='C1')

ax.text(1710,n_biotite/2,'ECMB: Ar-Ar biotite (~330Â°C)',size='11',horizontalalignment='left',
         verticalalignment='center',multialignment='center',color='C0')

ax.text(1710,-7,'pre-ECMB: Ar-Ar biotite (~330Â°C)',size='11',horizontalalignment='left',
         verticalalignment='center',multialignment='center',color='black')

ax.add_patch( Rectangle((1776.76,-12), 
                        4.68, 50, 
                        color='grey',alpha=0.5) )

plt.arrow(1770, 35, 5, -0.00005, length_includes_head=True,
          head_width=0.5, head_length=2,color='grey')

ax.text(1769,35.75,'time of\ndiabase dike\nemplacement',size='11',horizontalalignment='left',
         verticalalignment='top',multialignment='left',color='grey')

plt.xlim(1875,1600)
plt.ylim(-12,n_biotite+n_hornblende+n_apatite+n_zircon+n_zircon_new+0.5)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1],loc='upper center',bbox_to_anchor=(0.5, 1.18),framealpha=1,ncol=3,columnspacing=0.9,fontsize=9)
ax.set_yticks([])
plt.xlabel("Age (Ma)",fontsize=13)
plt.savefig('../Manuscript/Figures/ECMB_dates_thermochron.pdf')
```
