# ECMB low-temperature rock magnetic experiments data analysis

This notebook imports and visualizes data from low-temperature remanence rock magnetic experiments that were conducted on specimens from the northeast-trending dikes of the East Central Minnesota Batholith. These experiments were conducted in January 2020 at the Institute for Rock Magnetism by Nick Swanson-Hysell and Yiming Zhang.

This Jupyter notebook accompanies the paper:

Swanson-Hysell et al. 2021 **The paleogeography of Laurentia in its early years: new constraints from the Paleoproterozoic East Central Minnesota batholith**

## Import scientific python libraries

```python
import pmagpy.ipmag as ipmag
import pmagpy.pmag as pmag
import pandas as pd
pd.set_option('display.max_columns', 500)

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
```

## Import data

```python
NED2_8c_MPMS = pd.read_csv('../Data/rockmag_data/NED2_8c_MPMS.csv')
NED2_8c_MPMS_FC = NED2_8c_MPMS[NED2_8c_MPMS.experiment == 'FC']
NED2_8c_MPMS_ZFC = NED2_8c_MPMS[NED2_8c_MPMS.experiment == 'ZFC']
NED2_8c_MPMS_RTSIRM_cool = NED2_8c_MPMS[NED2_8c_MPMS.experiment == 'RTSIRM_cooling']
NED2_8c_MPMS_RTSIRM_warm = NED2_8c_MPMS[NED2_8c_MPMS.experiment == 'RTSIRM_warming']

NED4_1c_MPMS = pd.read_csv('../Data/rockmag_data/NED4_1c_MPMS.csv')
NED4_1c_MPMS_FC = NED4_1c_MPMS[NED4_1c_MPMS.experiment == 'FC']
NED4_1c_MPMS_ZFC = NED4_1c_MPMS[NED4_1c_MPMS.experiment == 'ZFC']
NED4_1c_MPMS_RTSIRM_cool = NED4_1c_MPMS[NED4_1c_MPMS.experiment == 'RTSIRM_cooling']
NED4_1c_MPMS_RTSIRM_warm = NED4_1c_MPMS[NED4_1c_MPMS.experiment == 'RTSIRM_warming']

NED18_2c_MPMS = pd.read_csv('../Data/rockmag_data/NED18_2c_MPMS.csv')
NED18_2c_MPMS_FC = NED18_2c_MPMS[NED18_2c_MPMS.experiment == 'FC']
NED18_2c_MPMS_ZFC = NED18_2c_MPMS[NED18_2c_MPMS.experiment == 'ZFC']
NED18_2c_MPMS_RTSIRM_cool = NED18_2c_MPMS[NED18_2c_MPMS.experiment == 'RTSIRM_cooling']
NED18_2c_MPMS_RTSIRM_warm = NED18_2c_MPMS[NED18_2c_MPMS.experiment == 'RTSIRM_warming']

NED34_6c_MPMS = pd.read_csv('../Data/rockmag_data/NED34_6c_MPMS.csv')
NED34_6c_MPMS_FC = NED34_6c_MPMS[NED34_6c_MPMS.experiment == 'FC']
NED34_6c_MPMS_ZFC = NED34_6c_MPMS[NED34_6c_MPMS.experiment == 'ZFC']
NED34_6c_MPMS_RTSIRM_cool = NED34_6c_MPMS[NED34_6c_MPMS.experiment == 'RTSIRM_cooling']
NED34_6c_MPMS_RTSIRM_warm = NED34_6c_MPMS[NED34_6c_MPMS.experiment == 'RTSIRM_warming']

NED36_8c_MPMS = pd.read_csv('../Data/rockmag_data/NED36_8c_MPMS.csv')
NED36_8c_MPMS_FC = NED36_8c_MPMS[NED36_8c_MPMS.experiment == 'FC']
NED36_8c_MPMS_ZFC = NED36_8c_MPMS[NED36_8c_MPMS.experiment == 'ZFC']
NED36_8c_MPMS_RTSIRM_cool = NED36_8c_MPMS[NED36_8c_MPMS.experiment == 'RTSIRM_cooling']
NED36_8c_MPMS_RTSIRM_warm = NED36_8c_MPMS[NED36_8c_MPMS.experiment == 'RTSIRM_warming']
NED36_8c_MPMS_RTSIRM_warm = NED36_8c_MPMS_RTSIRM_warm[NED36_8c_MPMS_RTSIRM_warm['reg fit'] > .95]
```

## Samples dominated by mid-coercivity component

```python
plt.figure(figsize=(9,6))
plt.subplot(2,2,1)
plt.plot(NED18_2c_MPMS_FC['T [K]'],NED18_2c_MPMS_FC['M [Am2/kg]'],'.-',color='C1',label='FC')
plt.plot(NED18_2c_MPMS_ZFC['T [K]'],NED18_2c_MPMS_ZFC['M [Am2/kg]'],'.-',color='C2',label='ZFC')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.legend()
plt.title('NED18-2c')

plt.subplot(2,2,3)
plt.plot(NED18_2c_MPMS_RTSIRM_cool['T [K]'],NED18_2c_MPMS_RTSIRM_cool['M [Am2/kg]'],'.-',color='C0',label='RTSIRM cooling')
plt.plot(NED18_2c_MPMS_RTSIRM_warm['T [K]'],NED18_2c_MPMS_RTSIRM_warm['M [Am2/kg]'],'.-',color='C3',label='RTSIRM warming')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.xlabel('T (K)')
plt.legend()

plt.subplot(2,2,2)
plt.plot(NED34_6c_MPMS_FC['T [K]'],NED34_6c_MPMS_FC['M [Am2/kg]'],'.-',color='C1',label='FC')
plt.plot(NED34_6c_MPMS_ZFC['T [K]'],NED34_6c_MPMS_ZFC['M [Am2/kg]'],'.-',color='C2',label='ZFC')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.legend()
plt.title('NED34-6c')

plt.subplot(2,2,4)
plt.plot(NED34_6c_MPMS_RTSIRM_cool['T [K]'],NED34_6c_MPMS_RTSIRM_cool['M [Am2/kg]'],'.-',color='C0',label='RTSIRM cooling')
plt.plot(NED34_6c_MPMS_RTSIRM_warm['T [K]'],NED34_6c_MPMS_RTSIRM_warm['M [Am2/kg]'],'.-',color='C3',label='RTSIRM warming')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.xlabel('T (K)')
plt.legend(loc='upper left')
plt.tight_layout()
```

## Samples dominated by high-coercivity component

```python
plt.figure(figsize=(9,6))
plt.subplot(2,2,1)
plt.plot(NED36_8c_MPMS_FC['T [K]'],NED36_8c_MPMS_FC['M [Am2/kg]'],'.-',color='C1',label='FC')
plt.plot(NED36_8c_MPMS_ZFC['T [K]'],NED36_8c_MPMS_ZFC['M [Am2/kg]'],'.-',color='C2',label='ZFC')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.legend()
plt.title('NED36-8c')

plt.subplot(2,2,3)
plt.plot(NED36_8c_MPMS_RTSIRM_cool['T [K]'],NED36_8c_MPMS_RTSIRM_cool['M [Am2/kg]'],'.-',color='C0',label='RTSIRM cooling')
plt.plot(NED36_8c_MPMS_RTSIRM_warm['T [K]'],NED36_8c_MPMS_RTSIRM_warm['M [Am2/kg]'],'.-',color='C3',label='RTSIRM warming')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.xlabel('T (K)')
plt.legend()

plt.subplot(2,2,2)
plt.plot(NED4_1c_MPMS_FC['T [K]'],NED4_1c_MPMS_FC['M [Am2/kg]'],'.-',color='C1',label='FC')
plt.plot(NED4_1c_MPMS_ZFC['T [K]'],NED4_1c_MPMS_ZFC['M [Am2/kg]'],'.-',color='C2',label='ZFC')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.legend()
plt.title('NED4-1c')

plt.subplot(2,2,4)
plt.plot(NED4_1c_MPMS_RTSIRM_cool['T [K]'],NED4_1c_MPMS_RTSIRM_cool['M [Am2/kg]'],'.-',color='C0',label='RTSIRM cooling')
plt.plot(NED4_1c_MPMS_RTSIRM_warm['T [K]'],NED4_1c_MPMS_RTSIRM_warm['M [Am2/kg]'],'.-',color='C3',label='RTSIRM warming')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.xlabel('T (K)')
plt.legend()
plt.tight_layout()
plt.show()
```

## Sample with dominantly a mid-coercivity component and a minor high-coercivity component

```python
plt.figure(figsize=(9,6))
plt.subplot(2,2,1)
plt.plot(NED2_8c_MPMS_FC['T [K]'],NED2_8c_MPMS_FC['M [Am2/kg]'],'.-',color='C1',label='FC')
plt.plot(NED2_8c_MPMS_ZFC['T [K]'],NED2_8c_MPMS_ZFC['M [Am2/kg]'],'.-',color='C2',label='ZFC')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.legend()
plt.title('NED2-8c')

plt.subplot(2,2,3)
plt.plot(NED2_8c_MPMS_RTSIRM_cool['T [K]'],NED2_8c_MPMS_RTSIRM_cool['M [Am2/kg]'],'.-',color='C0',label='RTSIRM cooling')
plt.plot(NED2_8c_MPMS_RTSIRM_warm['T [K]'],NED2_8c_MPMS_RTSIRM_warm['M [Am2/kg]'],'.-',color='C3',label='RTSIRM warming')
plt.xlim(0,300)
plt.ylabel('M (Am2/kg)')
plt.xlabel('T (K)')
plt.legend()
plt.show()
```
